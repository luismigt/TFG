# Decimal_coordinates_1.0

En este script se va a tratar de trasnformar el formato XX-YY-ZZ W/E a formato decimal, ya que entendemos que nos simplificar√° el trabajo a posteriori.

```python
import pandas as pd
import numpy as np

df_estaciones=pd.read_csv('../DATASETS/ESTACIONES/estaciones_coordenadas_2.csv',sep='|',error_bad_lines=False)
```

```python
df_estaciones.head()
```

```python
df_estaciones.info()
```

```python
#Todas las estaciones se encuentran en latitud NORTE, con lo cual, podemos eliminar la N sin ningun problema
df_estaciones['Latitud'].replace('N','', inplace=True, regex=True)
```

```python
df_estaciones.head()
```

```python
#Las estaciones se encuentran indistintamente a Este y Oeste del meridiano cero, con lo cual, a las del este las marcamos con un 
#False, y a las del Oeste con un True. 

df_estaciones['Flag']=df_estaciones['Longitud'].str.contains('W', regex=False)
df_estaciones.head()
```

```python
df_estaciones.head()
```

```python
conditions = [
    (df_estaciones['Flag']=='True'),
     (df_estaciones['Flag']=='False')]
choices=[df_estaciones['Longitud'].replace('W','', inplace=True, regex=True),
         df_estaciones['Longitud'].replace('E','', inplace=True, regex=True)]
```

```python
df_estaciones.head()
```

```python
df_estaciones[['Lat_grad','Lat_min','Lat_seg']]=df_estaciones.Latitud.str.split("-",expand=True)
df_estaciones[['Lon_grad','Lon_min','Lon_seg']]=df_estaciones.Longitud.str.split("-",expand=True)
df_estaciones.head()

```

```python
df_estaciones.fillna(value=pd.np.nan, inplace=True)
```

```python
df_estaciones.head()
```

```python
df_estaciones=df_estaciones.fillna(0)
```

```python
df_estaciones['Lat_grad'] = df_estaciones['Lat_grad'].astype(int)
df_estaciones['Lat_min'] = df_estaciones['Lat_min'].astype(int)
df_estaciones['Lat_seg'] = df_estaciones['Lat_seg'].astype(int)
df_estaciones['Lon_grad'] = df_estaciones['Lon_grad'].astype(int)
df_estaciones['Lon_min'] = df_estaciones['Lon_min'].astype(int)
df_estaciones['Lon_seg'] = df_estaciones['Lon_seg'].astype(int)
```

```python
df_estaciones.info()
```

```python
df_estaciones['Lat_seg'] = df_estaciones['Lat_seg']/60
df_estaciones['Lon_seg'] = df_estaciones['Lon_seg']/60
```

```python
df_estaciones.head()
```

```python
df_estaciones['Lat_min'] =df_estaciones['Lat_min']+df_estaciones['Lat_seg']
df_estaciones['Lon_min'] = df_estaciones['Lon_min']+df_estaciones['Lon_seg']
df_estaciones.head()
```

```python
df_estaciones['Lat_min'] =df_estaciones['Lat_min']/60
df_estaciones['Lon_min'] =df_estaciones['Lon_min']/60
df_estaciones['Lat_grad'] =df_estaciones['Lat_grad']+df_estaciones['Lat_min']
df_estaciones['Lon_grad'] = df_estaciones['Lon_grad']+df_estaciones['Lon_min']
df_estaciones.head()
```

```python
df_estaciones=df_estaciones.drop(['Lat_min'], axis=1)
df_estaciones=df_estaciones.drop(['Lat_seg'], axis=1)
df_estaciones=df_estaciones.drop(['Lon_min'], axis=1)
df_estaciones=df_estaciones.drop(['Lon_seg'], axis=1)
df_estaciones=df_estaciones.drop(['Latitud'], axis=1)
df_estaciones=df_estaciones.drop(['Longitud'], axis=1)
```

```python
df_estaciones.head()
```

```python
rslt_df = df_estaciones[df_estaciones['Flag'] ==True] 
```

```python
rslt_df.head()
```

```python
rslt_df2 = df_estaciones[df_estaciones['Flag'] ==False] 
```

```python
rslt_df2.head()
```

```python
rslt_df['Lon_grad']=rslt_df['Lon_grad']*(-1)
```

```python
rslt_df.head()
```

```python
df_estaciones=pd.concat([rslt_df,rslt_df2])
df_estaciones.head(n=250)
```

```python
df_estaciones=df_estaciones.drop(['Flag'], axis=1)
df_estaciones.head()
```

```python
df_estaciones.columns = ['ID','ESTACION','Estado','Latitud','Longitud']
```

```python
df_estaciones.head()
```

```python
df_estaciones.to_csv('../DATASETS/ESTACIONES/estaciones_coordenadas_decimal.csv',sep='|',header=True,index=False)
```

```python
df_estaciones['Longitud']=df_estaciones['Longitud'].astype(str)
df_estaciones['Latitud']=df_estaciones['Latitud'].astype(str)
```

```python
df_estaciones['Coordenadas']=df_estaciones[['Latitud','Longitud']].apply(lambda x: ', '.join(x), axis=1)
df_estaciones.info()
```

```python
df_estaciones.head(n=60)
```
