```python
import pandas as pd
import numpy as np
import statsmodels.api as sm


import seaborn as sn


import matplotlib.pyplot as plt


df=pd.read_csv('../../DATASETS/AHOGADOS/CSV/historico_inc_est_meteo_codigos_test.csv',sep='|',error_bad_lines=False)
```
