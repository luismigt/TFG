# Analiza_atributos_1.0

En este script analizamos como se comportan los atributos, a saber, que valores toman, como se repiten, etc

```python
import pandas as pd
import numpy as np

df_inc=pd.read_csv('../DATASETS/AHOGADOS/CSV/historico_incidentes_estaciones_meteo.csv',sep='|',error_bad_lines=False)
df_inc=df_inc.replace('Sin datos', np.NaN)
df_inc=df_inc.replace('Sin Datos', np.NaN)
#df_inc=df_inc[df_inc.CCAA != 'CANARIAS']
```

```python
df_inc.info()
```

```python
df_Nacionalidad=df_inc.groupby(['Nacionalidad']).size().to_frame(name="counts")
df_Nacionalidad=df_Nacionalidad.sort_values(by=['counts'],ascending=False).reset_index()
df_Nacionalidad['Rel']=df_Nacionalidad['counts']/6394*100
df_Nacionalidad.head(n=50)
#6394 
```

```python
df_Origen=df_inc.groupby(['Origen']).size().to_frame(name="counts")
df_Origen=df_Origen.sort_values(by=['counts'],ascending=False).reset_index()
#df_Origen['Mortal']=df_inc[df_inc.Pronostico = 'Ahogamiento mortal'].count() 
df_inc['Pronostico'].map(lambda x: x='Ahogamiento mortal').count()

df_Origen['Rel']=df_Origen['counts']/3813*100
df_Origen.head(n=99)
#3813
```

```python
#df_inc['Pronostico'].map(lambda x: x='Ahogamiento mortal').count()


df_inc.groupby('Origen').apply(lambda g: pd.Series(g['Origen'].unique()).str.startswith("T").sum())
```
