```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt


df=pd.read_csv('../../DATASETS/AHOGADOS/CSV/historico_inc_est_meteo_codigos_test.csv',sep='|',error_bad_lines=False)
```

```python
df_stats=df[['Mes', 'IdAhogado', 'IdPersona', 'Sexo', 'Edad', 'ID', 'Estacion', 'T_med', 'T_max', 'T_min', 'Precipitaciones', 'Presion','Dir_viento', 'V_Viento', 'Nubosidad', 'ProfNievecm', 'InsolacHoras','IdPronostico', 'Mortal','Costa']]
```

```python
df_stats['Sexo']=df_stats['Sexo'].str.replace('Hombre','0')
df_stats['Sexo']=df_stats['Sexo'].str.replace('Mujer','1')
df_stats['Sexo']=df_stats['Sexo'].str.replace('Sin Datos','2')
df_stats['Sexo']=df_stats['Sexo'].str.replace('NaN','2')

df_stats['Sexo'] = df_stats['Sexo'].fillna('2')
```

```python
#df_stats=df_stats[df_stats.Mortal == 1]


#df_stats = df_stats[(df_stats['Mortal'] == 1) & (df_stats['Costa'] == 0) & (df_stats['Mes'].between(6,9))]



df_stats_costa = df_stats[(df_stats['Mortal'] == 1) & (df_stats['Costa'] == 1)]# & (df_stats['Mes'].between(6,9))]
df_stats_nocosta = df_stats[(df_stats['Mortal'] == 1) & (df_stats['Costa'] == 0)]# & (df_stats['Mes'].between(6,9))]
```
